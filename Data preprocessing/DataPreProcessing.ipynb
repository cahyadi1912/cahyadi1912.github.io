```python
import pandas as pd
import os
```

```python
mortality_file= "Number of deaths children up to 5 years.csv"

mortality_df = pd.read_csv(mortality_file)
mortality_df = mortality_df[mortality_df["Dim1"] == "0-4 years"]
mortality_df = mortality_df[["SpatialDimValueCode", "Period", "Dim1", "Dim2", "FactValueNumeric"]]
mortality_df = mortality_df.rename(columns ={"SpatialDimValueCode": "CountryCode",
                                             "Period": "Year",
                                             "Dim1": "AgeGroup",
                                             "Dim2": "CauseOfDeath",
                                             "FactValueNumeric" : "Value"
                                            })

mortality_df = mortality_df.astype({"Year":"int64"})
    
```

```python
print(mortality_df)
print(mortality_df.dtypes)
```

```python
country_file = "WDICountry.csv"
country_df = pd.read_csv(country_file)
country_df = country_df.rename(columns = {"Country Code": "CountryCode",
                                          "Short Name": "CountryName",
                                          "Income Group": "IncomeGroup"
                                         })
```

```python
print(country_df)
```

```python
wdidata_file = "WDIData.csv"
gdp_df = pd.read_csv(wdidata_file)
gdp_df = gdp_df[gdp_df["Indicator Name"] == "GDP per capita (constant 2010 US$)"]
gdp_df = gdp_df[["Country Name", "Country Code", "2000", "2001", "2002", "2003", "2004", "2005",
                "2006", "2007", "2008","2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017"]]
gdp_df = gdp_df.rename(columns={"Country Name": "CountryName",
                                "Country Code": "CountryCode"})

```

```python
print(gdp_df)
```

```python
gdp_df = pd.melt(gdp_df, id_vars=["CountryName", "CountryCode"], 
                 value_vars= ["2000", "2001", "2002", "2003", "2004", "2005", "2006", "2007", "2008",
                              "2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017"])

gdp_df = gdp_df.rename(columns={"variable": "Year",
                                "value": "GDPPerCapita"
                               })

gdp_df = gdp_df.astype({"Year":"int64"})
```

```python
print(gdp_df)
print(gdp_df.dtypes)
```

```python
wdidata_file = "WDIData.csv"
population_df = pd.read_csv(wdidata_file)
population_df = population_df[population_df["Indicator Name"] == "Population, total"]
population_df = population_df[["Country Name", "Country Code", "2000", "2001", "2002", "2003", "2004", "2005",
                "2006", "2007", "2008","2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017"]]
population_df = population_df.rename(columns={"Country Name": "CountryName",
                                              "Country Code": "CountryCode"
                                             })

population_df = pd.melt(population_df, id_vars=["CountryName", "CountryCode"], 
                 value_vars= ["2000", "2001", "2002", "2003", "2004", "2005", "2006", "2007", "2008",
                              "2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017"])

population_df = population_df.rename(columns={"variable": "Year",
                                "value": "Population"
                               })
population_df = population_df.astype({"Year":"int64"})

```

```python
print(population_df)
print(population_df.dtypes)
```

```python
mergedmortality_df = mortality_df.merge(country_df[["CountryCode", "CountryName", "Region", "IncomeGroup"]], how="left", left_on='CountryCode', right_on='CountryCode')
print(mergedmortality_df)
print(mergedmortality_df.dtypes)
```

```python
mergedmortality_df = mergedmortality_df.merge(gdp_df[["CountryCode", "Year", "GDPPerCapita"]], how="left", 
                                              left_on=['CountryCode', 'Year'],  right_on=['CountryCode','Year'])
print(mergedmortality_df)
```

```python
mergedmortality_df = mergedmortality_df.merge(population_df[["CountryCode", "Year", "Population"]], how="left", 
                                              left_on=['CountryCode', 'Year'],  right_on=['CountryCode','Year'])
print(mergedmortality_df)
```

```python
mergedmortality_df = mergedmortality_df[ mergedmortality_df["Region"].notna()]
```

```python
mergedmortality_df.to_csv("childmortality.csv", index=False)
```

```python
mortalitycountry_df = mergedmortality_df.groupby(["CountryCode", "Year"], as_index=False).agg({"Value": "sum",
                                                                                  "GDPPerCapita": "min",
                                                                                  "Population" : "min"
                                                                                 })

mortalitycountry_df = mortalitycountry_df.merge(country_df[["CountryCode", "CountryName", "Region", "IncomeGroup"]], 
                                               how="left", left_on='CountryCode', right_on='CountryCode')

print(mortalitycountry_df)
print(mortalitycountry_df.dtypes)
```

```python
mortalitycountry_df.to_csv("childmortality_percountry.csv", index=False)
```

```python
causeofdeath_df = mergedmortality_df.groupby(by=["CauseOfDeath", "Year"], as_index=False).agg({"Value": "sum"})
print(causeofdeath_df)
causeofdeath_df.to_csv("causeofdeath_total.csv", index=False)
```

```python
toptencauseofdeath_df = mergedmortality_df.sort_values(["CauseOfDeath", "Year", "Value"], ascending=[True, True, False]);
print(toptencauseofdeath_df)
```

```python
counts = toptencauseofdeath_df.groupby(by=["CauseOfDeath", "Year"]).cumcount()
toptencauseofdeath_df["Rank"] =  counts
toptencauseofdeath_df = toptencauseofdeath_df[toptencauseofdeath_df["Rank"] < 10];
toptencauseofdeath_df.to_csv("toptencountries_causeofdeath.csv", index=False)
```
